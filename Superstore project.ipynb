```python
def description(data):
    '''
    Returns the DataFrame representing a thorough description of the data.
    '''
    dtypes = data.dtypes
    counts = data.apply(lambda col: col.count())
    nulls = data.apply(lambda col: col.isnull().sum())
    nan_ratio = nulls / counts
    uniques = data.apply(lambda col: col.unique())
    n_uniques = data.apply(lambda col: col.nunique())
    maxs = data.apply(lambda col: col.max())
    mins = data.apply(lambda col: col.min())
    
    cols = {'dtypes':dtypes, 'counts':counts, 'nulls' : nulls, 'nan_ratio': nan_ratio,
            'max':maxs, 'min':mins,'n_uniques':n_uniques, 'uniques':uniques}
    return pd.DataFrame(data=cols)
description(Salesdata)
```

```python
#REGION
# Compute the counts for each Region
Region_counts = Salesdata['Region'].value_counts()

print(Salesdata['Region'].value_counts())
# Create a countplot with the counts order
plt.figure(figsize=(5, 6))
plt.title('Count of Orders by Region', fontsize=20)
sns.countplot(x='Region', data=Salesdata, order=Region_counts.index, hue='Region', edgecolor='black', linewidth = 1.5)
plt.show()
```

```python
##STATE
# Compute the counts for each State
state_counts = Salesdata['State'].value_counts()

print ('Top 10 States with the higest counts of Orders;')
print(Salesdata['State'].value_counts()[:10])
# Create a countplot with the counts order
plt.figure(figsize=(20, 10))
plt.title('Count of Orders by State', fontsize=20)
sns.countplot(y='State', data=Salesdata, order=state_counts.index, hue='State', palette="Set2")
plt.show()
```

```python
## STATES WITH THE HIGHEST TOTAL REVENUE

top_States = Salesdata.groupby('State')['Sales'].sum()
top_States = pd.DataFrame(top_States).round(2).sort_values('Sales', ascending = False).reset_index()

#Let's change the index label
new_index_cities =[]
for i in range(1, len(top_States)+1):
    new_index_cities.append(i)   

#Let's apply the new indexes to the DF
top_States= top_States.set_index(pd.Index(new_index_cities))
top_States.head()
```

```python

plt.figure(figsize = (20,10))
ax = sns.barplot(x='State', y='Sales', data=top_States[:20], palette= 'Set1', hue='State', edgecolor='black', linewidth = 1.5)

plt.title('States which generated highest Revenue', fontsize=20)
plt.ylabel('Total Revenue $')
plt.xlabel('State')

for index, revenue in top_States["Sales"][:20].items():
    if revenue>350000:
        plt.text(index -1,revenue - 80000,str(revenue) + ' $', fontsize=12,rotation=90,color='k', horizontalalignment='center');
    else:
        plt.text(index-1,revenue + 10000, str(revenue) + ' $', fontsize=12,rotation=90,color='k', horizontalalignment='center');
plt.show()
```

```python
#datetime preprocessing
Salesdata['Order Date'] = pd.to_datetime(Salesdata['Order Date'], format='%d/%m/%Y')
Salesdata['Ship Date'] = pd.to_datetime(Salesdata['Ship Date'], format='%d/%m/%Y')
description(Salesdata)
```

```python
# Empty Postal Code preprocessing
# print out rows  with nan values
nan_rows = Salesdata.isna().any(axis=1)
print("\nRows with Postal Code holding NaN values:")
print(Salesdata[nan_rows].head(11))
```

It has been observed that all the data with missing postal code are in Burlington City.

Looking up the Postal Code fro Burlington, Vermont we have decide to fill the NaN values with the default postal code '05401'.

```python
# filling NaN values in Postal Code column

Salesdata['Postal Code'].fillna(05401.0, inplace=True)
```

```python
#Convert values in Postal code column to integers.
Salesdata['Postal Code'] = Salesdata['Postal Code'].astype(int)
```

```python
#Check to see if change has been implemented correctly
#Show all row with Burlington in 'City' column
Salesdata.loc[Salesdata['City'] == 'Burlington']
```
